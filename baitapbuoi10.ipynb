```python
#W9A1
def merge(a,b):
  if not a:
    return b
  if not b:
    return a
  if a[0] <= b[0]:
    return [a[0]]+ merge(a[1:],b)
  else:
    return [b[0]]+merge(a,b[1:])

a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(*merge(a,b))
```

```python
#W9A3
def min_after_one_swap(n):
  d = list(str(n))
  l=len(d)
  for i in range(l):
    k=i
    for j in range(i+1,l):
      if d[j] < d[k] and k!=0 and d[k]!=0:
        k=j
    if k != i:
      d[i],d[k]=d[k],d[i]
      break
  return str("".join(d))
n=int(input())
print(min_after_one_swap(n))
```

```python
#W9A4
def can_convert(a,b):
  j=0
  for ch in a:
    if j < len(b) and ch.upper() == b[j]:
      j+=1
    elif ch.isupper():
      return "NO"
  return "YES" if j == len(b) else "NO"
Q=int(input())
for _ in range(Q):
  a=input().strip()
  b=input().strip()
print(can_convert(a,b))
```

```python
#W9A5
def is_subarray(list1,list2):
  n,m=len(list1), len(list2)
  if n>m:
    return False
  for i in range(m-n+1):
    if list2[i:i+n] == list1:
      return True
  return False
list1=list(map(int,input().split()))
list2=list(map(int,input().split()))
if is_subarray(list1,list2):
  print("YES")
else:
  print("NO")
```

```python
#W9A8
def is_path_crossing(moves):
  x,y=0,0
  visited = set()
  visited.add((0,0))
  for i in moves:
    if i == 'R':
      x+=1
    elif i == 'L':
      x-=1
    elif i == 'U':
      y+=1
    elif i == 'D':
      y-=1
    if (x,y) in visited:
      return True
    visited.add((x,y))
  return False
moves=input()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
  a,b,c,d,e=nums
  if a>b:
    a,b=b,c
  if c>d:
    c,d=d,c
  if a>c:
    a,c=c,a
    b,d=d,b
  if b>c:
    b,c=c,b
  if e<b:
    return b
  elif e>c:
    return c
  else:
    return e
nums=list(map(int,input().split()))
print(Median(nums))
```

```python
#W9A10
def findRadius(houses,heaters):
  houses.sort()
  heaters.sort()
  i = 0
  r = 0
  for h in houses:
    while i < len(heaters) -1 and abs(heaters[i+1]-h) <= abs(heaters[i]-h):
      i+=1
    r = max(r,abs(heaters[i] - h))
  return r
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses,heaters))
```
