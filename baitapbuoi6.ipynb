```python
#W6A9
n = eval(input())
m = eval(input())
d={}
for k in n:
  d[k]=n[k]
for k in m:
  if k in d:
    d[k]+=m[k]
  else:
    d[k]=m[k]
kq={}
for k in sorted(d):
  kq[k]=d[k]
print(kq)
```

```python
#W6A10
m=list(map(int,input().split()))
n=int(input())
kq=[]
a=len(m)
for i in range(a):
  for j in range(i+1,a):
    if m[i] + m[j] == n:
      kq.append((m[i],m[j]))
kq.sort(key= lambda x:x[0])
print(kq)
```

```python
#W6A12
chuoi=list(map(int,input().split()))
dem={}
for i in chuoi:
  if i in dem:
    dem[i]+=1
  else:
    dem[i]=1
max_dem=0
for i in dem:
  if dem[i] >max_dem:
    max_dem=dem[i]
kq=None
for i in dem:
  if dem[i]== max_dem:
    if kq is None or i < kq:
      kq=i
print(kq)
```
