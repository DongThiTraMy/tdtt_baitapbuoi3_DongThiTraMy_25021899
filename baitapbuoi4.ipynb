```python
#W4A7
n=int(input("Nhập số nguyên dương n (n>=2): "))
def so_nguyen_to(a):
  if a<2:
    return False
  for i in range(2,a):
    if a%i==0:
      return False
  return True
def uoc_lon_nhat(c):
  for i in range(c,1,-1):
    if c%i==0 and so_nguyen_to(i):
      return i
print("Ước số nguyên lớn nhất và là số nguyên tố là: ",uoc_lon_nhat(n))
```

```python
#W4A8
def reverse_number(n):
    new = 0
    while n != 0:
        new = new * 10 + n % 10
        n //= 10
    return new
def so_dxung(n):
    buoc = 0
    while n != reverse_number(n):
        n = n + reverse_number(n)
        buoc += 1
    print(buoc, n, sep=' ')
a= int(input())
so_dxung(a)
```

```python
#W4A9
n=int(input("Nhập vào số nguyên dương n: "))
def distinct_digits(x):
  s=str(x)
  return len(set(s))==len(s)
ket_qua=[]
i=1
while i*i<=n:
  a=i*i
  if distinct_digits(a):
    ket_qua.append(str(a))
  i=i+1
if ket_qua:
  print(" ".join(ket_qua))
else:
  print("no number")
```

```python
#W4A10
def collatz_length(x):
    count = 1
    while x!=1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        count += 1
    return count

n = int(input("Nhập số nguyên dương n: "))
if n <= 0:
    print("Vui lòng nhập số nguyên dương.")
else:
    max_len = 0
    so_nho_nhat = 1

    for i in range(1, n + 1):
        length = collatz_length(i)
        if length > max_len:
            max_len = length
            so_nho_nhat = i
        elif length == max_len and i < so_nho_nhat:
            so_nho_nhat = i

    print(so_nho_nhat, max_len)

```

```python
#W4A14
def uoc_chung_lon_nhat(m, n):
    while n != 0:
        m, n = n, m % n
    return m

m = int(input("Nhập số m: "))
n = int(input("Nhập số n: "))

if m <= 0 or n <= 0:
    print("Vui lòng nhập số nguyên dương.")
else:
    print("Ước chung lớn nhất là:", uoc_chung_lon_nhat(m, n))
```

```python
#W4A26
def fibonacci(n):
  f1,f2=1,1
  while f2 <= n:
    f1,f2=f2,f1+f2
  return f1
n=int(input())
print(fibonacci(n))
```
