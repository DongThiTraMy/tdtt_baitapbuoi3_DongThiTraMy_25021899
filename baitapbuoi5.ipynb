```python
#W5A3
import math
def so_nguyen_to(n):
  if n<2:
    return False
  for i in range(2,math.isqrt(n)+1):
    if n % i == 0:
      return False
  return True
a=int(input())
if so_nguyen_to(a):
  print("True")
else:
  print("False")
```

```python
#W5A4
def so_hoan_hao(n):
  tong=0
  for i in range(1,n):
    if n%i==0:
      tong+=i
  return tong==n
a=int(input())
if so_hoan_hao(a):
  print("True")
else:
  print("False")
```

```python
#W5A6
def may_tinh(num1: float, num2: float, operat: str):
    if operat == '+':
        kq = num1 + num2
    elif operat == '-':
        kq = num1 - num2
    elif operat == '*':
        kq = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: chia cho 0!"
        kq = num1 / num2
    else:
        return "Toán tử không hợp lệ!"
    return kq
a = float(input("Nhập số thứ nhất: "))
b = float(input("Nhập số thứ hai: "))
op = input("Nhập toán tử (+ - * /): ")

print(f"{may_tinh(a, b, op):.2f}")
```

```python
#W5A9
n=int(input())
def tong(n):
  tong=0
  while n>0:
    tong+=n%10
    n=n//10
  return tong
print(tong(n))

```

```python
#W5A10
def dang_cau(a,b):
  mapping={}
  mapped_values = set()
  for ca, cb in zip(a,b):
    if ca in mapping:
      if mapping[ca] != cb:
        return False
    else:
      if cb in mapped_values:
        return False
      mapping[ca] = cb
      mapped_values.add(cb)
  return True
a,b=map(str,input().split())
if dang_cau(a,b):
  print("True")
else:
  print("False")
```
